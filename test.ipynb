```python
from flask import Flask, request, jsonify
import redis
import uuid

app = Flask(__name__)

# Redis 초기화
r = redis.Redis(host='localhost', port=6379, db=0)

# 일정 생성 및 임시 저장
@app.route('/create_schedule', methods=['POST'])
def create_schedule():
    data = request.json
    schedule_id = str(uuid.uuid4())  # 고유한 ID 생성
    r.hmset(schedule_id, data)  # Redis에 일정 저장
    r.expire(schedule_id, 600)  # 10분 후 만료 (임시 저장)
    return jsonify({"message": "Schedule created!", "schedule_id": schedule_id})

# 일정 등록 및 DB 저장 (여기서는 예제로 Redis만 사용)
@app.route('/register_schedule', methods=['POST'])
def register_schedule():
    schedule_id = request.json.get('schedule_id')
    if not r.exists(schedule_id):
        return jsonify({"message": "Invalid schedule ID or expired!"}), 400
    # 실제 DB에 저장 (이 예제에서는 Redis만 사용하므로 생략)
    # DB 저장 후 Redis에서 해당 정보 삭제
    r.delete(schedule_id)
    return jsonify({"message": "Schedule registered successfully!"})

if __name__ == '__main__':
    app.run(debug=True)
```

```python
import re

def parse_schedule(text):
    # 정규 표현식을 사용하여 표 찾기
    rows = text.strip().split("\n")[2:]  # 첫 두 줄은 헤더와 구분선이므로 제외
    parsed_rows = [list(map(str.strip, filter(None, row.split("|")))) for row in rows]
    return parsed_rows

text = """
| 일자 | 시간 | 여행지 | 설명 |
|------|------|--------|------|
| 9/23 | 10:00-12:00 | 서울 명동 | 명동 쇼핑 및 길거리 음식 체험 |
| 9/23 | 14:00-16:00 | 서울 남산타워 | 타워 방문 및 도심 전망 감상 |
"""

schedules = parse_schedule(text)
print(schedules)
```
